```python
import json
```

```python
# Laod the json file

def load_data(filename):
    with open(filename,"r") as file:
        data=json.load(file)
    return data

data=load_data("data.json")
print(type(data))
print(data.keys()) 
```

```python
def display_users(data):
    print("Users and their connections:\n")
    for user in data["users"]:
        print(f"ID:{user['id']} {user['name']} and their friends are {user['friends']} and they liked the pages {user['liked_pages']}.")
    print("\nPages:\n")
    for page in data["pages"]:
        print(f"{page['id']}: {page['name']}")
display_users(data)
```
